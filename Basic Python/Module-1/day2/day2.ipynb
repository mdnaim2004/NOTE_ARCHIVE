### String Concatination
add two or Multiple String in use (+) 
    example: 

            s1 = "Hello,"

            s2 = "World!"

            add_string = s1 + s2

            Hello,World!

```python
s1 = "Hello,"
s2 = "World!"
add_string = s1 + s2

print(add_string)
```
